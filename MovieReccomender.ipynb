```python
import pandas as pd
import numpy as np
```

```python
cols=['User_ID','Movie_ID','Rating']
dataFile=pd.read_csv("u.data",sep="\t", names=cols,usecols=range(3))

movieCols=['Movie_ID','Movie_Title','Release_Date','Video_Release_Date','IMDb_URL','Unknown','Action','Adventure','Animation','Children','Comedy','Crime','Documentary','Drama','Fantasy','Film_Noir','Horror','Musical','Mystery','Romance','Sci-Fi','Thriller','War','Western']
mvCols=['Movie_ID','Title']
movies=pd.read_csv('u.item',sep="|", names=mvCols,usecols=range(2),encoding='latin-1')

ratings=pd.merge(movies,dataFile)
```

```python
ratings
```

```python
ratings.reset_index(drop=True,inplace=True)
```

```python
ratings
```

```python
movieRatings=ratings.pivot_table(index=['User_ID'],columns=['Title'],values='Rating')

movieRatings
```

```python
starWarsRating=movieRatings['Star Wars (1977)']
starWarsRating
```

```python
similarSWMovie=movieRatings.corrwith(starWarsRating)
similarSWMovie.dropna(inplace=True)
similarSWMovies=pd.DataFrame(similarSWMovie)

similarSWMovies
```

```python
similarSWMovies.sort_values(ascending=False,by=0)
```

```python
movieStats=ratings.groupby('Title').agg({'Rating': [np.size, np.mean]})

movieStats
```

```python
popularMovies=movieStats['Rating']['size'] >=100
movieStats[popularMovies].sort_values([('Rating','mean')],ascending=False)[:15]
```

```python
data=movieStats['Rating'].join(pd.DataFrame(similarSWMovies,columns=['similarity']))

data
```
